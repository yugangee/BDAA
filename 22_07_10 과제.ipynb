```python
import pandas as pd
```

```python
exam = pd.read_csv('exam.csv')
```

```python
# 과제1 1반인데 수학은 50 이하, 영어 80이상인 학생 추출
exam.query('nclass == 1 & math <= 50 & english >= 80')
```

```python
# 과제2 오름차순, 내림차순
```

```python
exam.sort_values('english') #오름차순
```

```python
exam.sort_values('math') # 오름차순
```

```python
exam.sort_values('science', ascending =False) #과학 점수를 기준으로 내림차순을 하는 경우
```

```python
exam.sort_values(['math','english']) #수학->영어 오름차순
```

```python
exam.sort_values(['math','english'], ascending = [True, False] ) #오름차순, 내림차순
```
