## 시계열데이터 컬럼을 인덱스로 바꾸기!

```python
crime_df =crime.set_index('REPORTED_DATE')
```

### 시계열데이터를 슬라이싱의 개념!

```python
crime_df
```

```python
#시계열데이터를 인덱스로 설정하면 여러가지 사용할 수 있는 문법이 있다.
crime_df.loc['2014']
```

```python
crime_df.loc['2014-05']
```

```python
crime_df.loc['2014-05-01']
```

```python
crime_df.loc['Dec 2014']
```

```python
crime_df.loc['Dec 2014'].sort_index()#인덱스 정렬
```

```python
#Datetime index만 작동하는 메서드!
crime_df.between_time('3:00','6:00',include_end=False)
```

```python
##내가 원하는 시간 데이터를 추출할 수 있다!
crime_df.at_time('9:18').sort_index()
```

```python
crime_df_1=crime_df.sort_index()
```

```python
crime_df_1.first(pd.offsets.MonthEnd(5))
#first 문법 구글링을 통해서 다양한 메서드를 찾아볼 수 있습니다!
```

```python
crime_df_1.first('5D') #5일을 기준으로 뽑기
```

### 시계열 데이터의 Datetime 다양한 문법을 구글링해서 11월에 발표해 주세요!

- resample을 이용하여 주별, 월별 값 계산하기!

```python
crime_df_1.resample('W').sum()
```

```python
crime_df_1.resample('W').mean()
```

```python
crime_df_1.resample('W').median()
```

```python
crime_df_1.resample('W').min()
```

```python
crime_df_1.resample('M').sum() #월별
```

```python
crime_df_1.resample('Y').sum() #월별
```
