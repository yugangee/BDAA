```python
import pandas as pd
df = pd.read_csv('재무재표샘플데이터.csv')
df
```

```python
#columns
df.columns
```

```python
df.기업.value_counts()
```

```python
#중복 제거한 기업의 수
df['기업'].drop_duplicates().to_frame() #중복을 제거한 실제 기업 수
```

```python
df.isna().sum()
```

```python
#부채비율이 40 초과인 회사 카운트
df['부채비율>40']=df['부채비율'] >40
df['부채비율>40'].value_counts()
```

```python
#비율
df['부채비율>40'].value_counts(normalize=True)
```

```python
df['매출액_상태'].value_counts()
```

```python
#groupby 이용
df.groupby('기업')['현재_주가'].max()
```

```python
df.groupby(['기업','연도','부채비율>40'])['ROE'].max()
```

```python
import numpy as np
df.groupby('기업',as_index=False)[['연도','부채비율','매출액증가율']].agg({'부채비율':['max','min'],
                                                       '매출액증가율':[np.mean]})
```
