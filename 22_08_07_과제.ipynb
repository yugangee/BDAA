```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#선색깔 정하기
plt.plot([1,2,3],[4,5,6],color='skyblue', label='line1')
plt.plot([1,2,3],[5,6,3],color='pink', label='line2')#선 색상 변경
plt.legend(loc='upper right')
```

```python
#다이아몬드, 원 마커
plt.plot([1,2,3],[4,5,6], color='skyblue', marker = 'D', markersize= 10,label='line1')
plt.plot([1,2,3],[5,6,3], color='pink', marker = 'o', markersize= 11,label='line2')
plt.legend(loc='upper right')
```

```python
#점선
plt.plot([1,2,3],[4,5,6],color='skyblue',linestyle='dashed', label='line1')
plt.plot([1,2,3],[5,6,3],color='pink',linestyle='dashed', label='line2')#선 색상 변경
plt.legend(loc='upper right')
```

```python
#선 굵기 두껍게
plt.plot([1,2,3],[4,5,6],color='skyblue',linestyle='dashed',linewidth =4, label='line1')
plt.plot([1,2,3],[5,6,3],color='pink',linestyle='dashed',linewidth =4, label='line2')#선 색상 변경
plt.legend(loc='upper right')
```

```python
#whitegrid 사용하기
sns.set(style= 'whitegrid')
plt.plot([1,2,3],[4,5,6],color='skyblue',linestyle='dashed',linewidth =4, label='line1')
plt.plot([1,2,3],[5,6,3],color='pink',linestyle='dashed',linewidth =4, label='line2')#선 색상 변경
plt.legend(loc='upper right')
```

2.마커

```python
plt.plot([1,2,3],[4,5,6], color='skyblue', marker = '.', markersize= 10,label='point')
plt.plot([1,2,3],[5,6,3], color='pink', marker = ',', markersize= 11,label='pixel')
plt.legend(loc='upper right')
```

```python
plt.plot([1,2,3],[4,5,6], color='skyblue', marker = 'o', markersize= 10,label='circle')
plt.plot([1,2,3],[5,6,3], color='pink', marker = 'v', markersize= 11,label='triangle_down')
plt.plot([1,2,3],[1,3,4], color='purple', marker = '^', markersize= 11,label='triangle_down')
plt.legend(loc='upper right')
```

```python
plt.plot([1,2,3],[4,5,6], color='skyblue', marker = 's', markersize= 10,label='square')
plt.plot([1,2,3],[5,6,3], color='pink', marker = 'p', markersize= 11,label='pentagon')
plt.plot([1,2,3],[1,3,4], color='purple', marker = '*', markersize= 11,label='star')
plt.plot([1,2,3],[2,4,5], color='orange', marker = 'h', markersize= 11,label='hexagon')
plt.legend(loc='upper right')
```

3.linestyle

```python
plt.plot([1,2,3],[4,5,6],color='skyblue',linestyle='dashed',linewidth =4, label='dashed line style')
plt.plot([1,2,3],[5,6,3],color='pink',linestyle='solid',linewidth =4, label='solid line style')#선 색상 변경
plt.legend(loc='upper right')
```

```python
plt.plot([1,2,3],[4,5,6],color='skyblue',linestyle='-.',linewidth =4, label='dash-dot line style')
plt.plot([1,2,3],[5,6,3],color='pink',linestyle=':',linewidth =4, label='dotted line style')#선 색상 변경
plt.legend(loc='upper right')
```

4.color

```python
#b: blue
#g: green
#r: red
#c: cyan
plt.plot([1,2,3],[4,5,6], color='b', marker = 's', markersize= 10,label='square')
plt.plot([1,2,3],[5,6,3], color='g', marker = 'p', markersize= 11,label='pentagon')
plt.plot([1,2,3],[1,3,4], color='r', marker = '*', markersize= 11,label='star')
plt.plot([1,2,3],[2,4,5], color='c', marker = 'h', markersize= 11,label='hexagon')
plt.legend(loc='upper right')
```

```python
#m: magenta
#y: yellow
#k: black
#w: white
plt.plot([1,2,3],[4,5,6], color='m', marker = 's', markersize= 10,label='square')
plt.plot([1,2,3],[5,6,3], color='y', marker = 'p', markersize= 11,label='pentagon')
plt.plot([1,2,3],[1,3,4], color='k', marker = '*', markersize= 11,label='star')
plt.plot([1,2,3],[2,4,5], color='w', marker = 'h', markersize= 11,label='hexagon')
plt.legend(loc='upper right')
```

5.선 끝 모양

```python
plt.plot([1,2,3],[1,3,4], color='k', linestyle='dashed',linewidth=10, dash_capstyle='butt', markersize= 11,label='1')
plt.plot([1,2,3],[2,4,5], color='w', linestyle='dashed',linewidth=10, dash_capstyle='round', markersize= 11,label='2')
plt.legend(loc='upper right')
```
