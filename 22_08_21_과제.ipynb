```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'이름':['영희','철수','민지','길동'],
                  '혈액형':['A','A','O','B']})
df2 = pd.DataFrame({'name':['영희','철수','민지','길동'],
                  '키':[160,180,168,179]})
```

```python
display(df1)
display(df2)
```

```python
pd.merge(df1,df2, left_on='이름',right_on='name')
```
