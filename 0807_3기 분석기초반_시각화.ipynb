## 선그래프

```python
#plot함수를 이용해서 시각화!
plt.plot([1,2,3],[4,5,6], label='line1')
plt.plot([1,2,3],[5,6,3], label='line2')
plt.legend(loc='upper right')
```

### 색상 설정하는 법!

```python
plt.plot([1,2,3],[4,5,6],color='red', label='line1')
plt.plot([1,2,3],[5,6,3],color='black', label='line2')#선 색상 변경
plt.legend(loc='upper right')
```

### 선 스타일 및 두께 설정


```python
plt.plot([1,2,3],[4,5,6],color='red',linestyle='dashed',linewidth=2 ,label='line1')
plt.plot([1,2,3],[5,6,3],color='black',linestyle='dashed',linewidth =4, label='line2')#선 색상 변경
plt.legend(loc='upper right')
```

### 마커 설정!
-marker ".", "*", "s"(사각형), "D"(다이아몬드),"d"(홀쭉한 다이아몬드)

```python
#plot함수를 이용해서 시각화!
plt.plot([1,2,3],[4,5,6],marker = 's', markersize= 10,label='line1')
plt.plot([1,2,3],[5,6,3],marker = 'o', markersize= 11,label='line2')
plt.legend(loc='upper right')
```

## 산점도 마커 설정

```python
#scatter 산점도
plt.scatter([1,2,3],[4,5,6], marker= '*',s=40, color='red') #line1 #마커, 사이즈, 컬러
plt.scatter([1.5,2.5],[5,6], marker= 's',s=100, color='blue') #line2 #마커, 사이즈, 컬러
plt.show()
```

### 투명도

```python
x
```

- 이상치의 기준을 판단할때
1. 분석가나, 현직자가 주관적으로 판단해서 이상치를 제거할 수도 있고
2. 통계적인 기법을 통해서 이상치를 제거할 수 있다.( IQR )

```python
x = np.random.normal(10,3,100)
y = np.random.normal(10,2,200)
```

```python
plt.boxplot([x,y])
```
