# 데이터 병합, 데이터 연결

### Concat 

- 데이터프레임을 연결할 때 사용하는 경우!

- 테이블과 테이블을 덩어리채 합친다 ( 공통 컬럼을 기준으로!)

```python
import pandas as pd
```

```python
df1 =pd.read_csv('concat_1.csv')
df2 =pd.read_csv('concat_2.csv')
df3 =pd.read_csv('concat_3.csv')
```

```python
print(df1)
print(df2)
print(df3)
```

```python
ct_df = pd.concat([df1,df2,df3]) #순서에 따라서 데이터가 병합
```

```python
ct_df_1 = pd.concat([df1,df3,df2])
```

```python
ct_df
```

```python
new_df = pd.DataFrame([['n1','n2','n3','n4']],columns = ['A','B','C','D'])
```

```python
new_df
```

```python
pd.concat([df1, new_df],ignore_index = True)#인덱스를 재설정
```

```python
#열방향으로 데이터를 병합
col_df =pd.concat([df1,df2,df3], axis=1) #축을 1로 조정해서 붙인다.
```

```python
col_df['A']#공통적인 컬럼만 추출이 된다.
```

```python
col_df['E'] = ['e1','e2','e3','e4']
```

```python
#df 컬럼을 다르게 바꿔보자!
df2.columns=['C','F','G','H']
df3.columns=['I','J','A','C']
```

```python
pd.concat([df1,df2,df3]) #분석가 예상하지 못한 방향으로 병합을 하게 되면 이런식의 NaN 값이 존재
```

```python
pd.concat([df1,df2,df3],axis=1)
```

## Merge
- 데이터 프레임 병합의 개념
- join (교집합)

```python
site = pd.read_csv('survey_site.csv')
survey = pd.read_csv('survey_survey.csv')
visited = pd.read_csv('survey_visited.csv')
```

```python
site
```

```python
visited
```

```python
site.merge(visited, left_on='name',right_on='site')
#처음 테이블.merge(오른쪽에 붙을 테이블, left_on='왼쪽 공통 컬럼이름',right_on='오른쪽 공통 컬럼 이름')
```

```python
df_new2=visited.merge(site, left_on='site',right_on='name')
```
