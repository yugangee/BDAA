## 데이터 전처리에 필요한 스킬

- pandas 데이터를 전처리/ 필요한 문법

- - query(), 행 추출
- - df[], 열 추출
- - sort_values() 정렬
- - groupby() 집단별로 나누기
- - assign() 변수추가
- - agg() 통계치 구하기
- - merge() 데이터 합치기(열)
- - concat() 데이터 합치기(행)

```python
import pandas as pd
exam = pd.read_csv('exam.csv')
```

```python
exam.query('nclass == 1') # . 을 이용해서 문법을 건다.  '.'이용해서 문법을 확장할 수 있다.
```

```python
exam.query('math == 50') 
```

```python
exam.query('nclass != 1') #1반을 제외한 나머지 반 모두 출력
```

```python
#비교 연산자도 사용할 수 있음
#이상, 이하, 초과, 미만
exam.query('math > 50')
```

```python
exam.query('math < 50')
```

```python
exam.query('math <= 50')
```

```python
exam.query('math >= 50')
```

```python
#논리 연산를 이용해서 여러가지 행 추출 
exam.query('nclass == 2 & math >=50') #다양한 조건을 걸어서 데이터를 추출할 수 있다.
```

```python
exam.query('nclass ==2 | math >=80') #or 조건
```

```python
exam.query('nclass ==1 | nclass==2 | nclass==3') #or 조건을 이용한 다중 조건
```

###  파생변수 만들기!

```python
exam['mean'] = (exam['math']+exam['english']+exam['science'])/3
```

### numpy where() 이용해서 조건에 따른 값을 부여한 변수를 만들 수 있다.

```python
import numpy as np

exam.assign(test = np.where(exam['science'] >=50,'pass','fail'))
```

```python
np.where(exam['science'] >=50)
```

```python
#lambda 식을 이용해서 파생 변수 만들기
exam.assign(new = lambda x : x['math']+x['english']+x['science'])
```
