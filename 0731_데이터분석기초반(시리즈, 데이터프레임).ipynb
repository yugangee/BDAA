## 데이터 시각화

```python
import seaborn as sns #시각화에 필요한 모듈!
```

```python
df = sns.load_dataset('anscombe')
```

```python
df
```

```python
#matplotlib 라이브러리 모듈
%matplotlib notebook
import matplotlib.pyplot as plt
```

### 시각화는 원하는 값( 또는 보여주고 싶은 값)을 추출하고 그 데이터를 가지고 시각화를 진행하는 것이 더 좋다!
- 전달성 문제

```python
df_1=df[df['dataset']=='I']
```

```python
plt.plot(df_1['x'],df_1['y'])
```

```python
plt.plot(df_1['x'],df_1['y'],'o')
```

```python
df.dataset.value_counts()
```

```python
df_1=df[df['dataset']=='I']
df_2=df[df['dataset']=='II']
df_3=df[df['dataset']=='III']
df_4=df[df['dataset']=='IV']
```

```python
df_4
```

```python
#4개의 값을 시각화 처리 하는 경우!
#그래프를 그리기 위한 격자 기본틀 만들기

fig = plt.figure()
axes1 =fig.add_subplot(2,2,1)
axes2 =fig.add_subplot(2,2,2)
axes3 =fig.add_subplot(2,2,3)
axes4 =fig.add_subplot(2,2,4)

axes1.plot(df_1['x'], df_1['y'], 'o')
axes2.plot(df_2['x'], df_2['y'], 'o')
axes3.plot(df_3['x'], df_3['y'], 'o')
axes4.plot(df_4['x'], df_4['y'], 'o')
```
