# 데이터 이용하기

```python
sale = pd.read_csv('sale1.csv')
```


# 정보 출력


```python
sale.info()
```

# 특정 column pandas로 불러오기


```python
sale['InvoiceDate']=pd.to_datetime(sale['InvoiceDate'])
```

```python
sale['InvoiceDate']
```

# 시계열 데이터 column을 인덱스로 바꾸기

```python
sale_df =sale.set_index('InvoiceDate')
sale_df
```

# 인덱스 이용해서 모아보기  이름.loc['년-월-일/년-월/월(영어) 년]

```python
sale_df.loc['2011-12-01']
```

```python
sale_df.loc['2011-12-09']
```

```python
sale_df.loc['DEC 2011']
```

# 인덱스 정렬하기

```python
sale_df.loc['DEC 2011'].sort_index()
```

# 날짜 조건


```python
sale_df.loc['2011-12-07':'2011-12-08'].sort_index()
```

# Datetime index만 작동하는 메서드!
시간조건

```python
sale_df.between_time('8:00','09:00',include_end=False)
```

# 원하는 시간 추출

```python
sale_df.at_time('11:18').sort_index()
```

```python
sale_df_1=sale_df.sort_index()
```

```python
sale_df_1.first(pd.offsets.MonthEnd(5))
```

# 주별 값 계산

```python
sale_df_1.resample('M').median()
```
