## Concat 함수
- 행, 열이건 덩어리 형태로 붙이는 것
- axis ( 축을 설정 )
- ignore_index 병합 후 인덱스 재설정!

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'A':[1,2,3],'B':[4,5,6]})
df2 = pd.DataFrame({'A':[7,8],'B':[9,10]})
```

```python
print(df1)
print(df2)
```

```python
#데이터 병합하기
display(pd.concat([df1,df2], axis=0, ignore_index = True))
#인덱스 기준에 따라 달라진다!
display(pd.concat([df1,df2], axis=0, ignore_index = False))
```

```python
display(pd.concat([df1,df2], axis=1, ignore_index = True))
#인덱스 기준에 따라 달라진다!
display(pd.concat([df1,df2], axis=1, ignore_index = False))
```

## Merge 함수
- 두 개의 데이터프레임을 합치는데 ( join ) 공통된 컬럼을 기준으로 두 데이터프레임이 병합, 연결 된다.

- merge 함수의 주요 인자
- left : 병합할 첫 번째 데이터프레임으로, 왼쪽 데이터 프레임을 뜻함
- right : 병합할 두 번째 데이터프레임, 오른쪽 데이터 프레임
- on : 조인, 공통된 컬럼의 목록으로 여기에 입력된 컬럼은 두 데이터프레임에 모두 포함 
- left_on : 왼쪽 데이터프레임에서 join키로 사용할 컬럼의 목록
- right_on : 오른쪽 데이터프레임에서 Join키로 사용할 컬럼의 목록
- left_index: 왼쪽 데이터프레임엥서 조인 기준으로 인덱스를 사용할지 
- right_index: 오른쪽 데이터프레임엥서 조인 기준으로 인덱스를 사용할지 
- how : 조인 방법을 결정한다. 기본 디폴트는 'inner' 조인( 교집합 ) 

```python
df1 = pd.DataFrame({'이름':['영희','철수','민지','길동'],
                  '반':[1,2,3,4]})
```

```python
df2 = pd.DataFrame({'이름':['영희','철수','민지','길동'],
                  '등급':[1,1,3,3]})
```

```python
print(df1)
print(df2)
```

```python
#on 인자를 사용해서 왼쪽 데이터프레임와 오른쪽 데이터프레임의 공통인 조인 키를 설정 ( 이름 )
merged_df = pd.merge(df1,df2, on='이름')
```

```python
merged_df
```

### 조인 방법 결정: how
- inner(디폴트) : 왼쪽과 오른쪽 데이터프레임의 조인 키에 대한 교집합 기준으로 병합 
- left : 왼쪽 데이터프레임의 조인 키를 기준으로 병합
- right : 오른쪽 데이터프레임의 조인 키를 기준으로 병합
- outer : 왼쪽과 오른쪽 두 데이터프레임의 조인 키에 대한 합집합으로 기준을 병합

```python
df1 = pd.DataFrame({'이름':['영희','철수','민지','길동'],
                  '반':[1,2,3,4]})
df2 = pd.DataFrame({'이름':['광수','철수','민지','길동'],
                  '등급':[1,1,3,3]})
```

```python
display(df1)
display(df2)
```

```python
pd.merge(df1, df2, on='이름',how='inner')  #교집합만 가지고 온다!
```

```python
#left
pd.merge(df1,df2, on='이름',how='left')
```

```python
#right
pd.merge(df1,df2, on='이름',how='right')
```

```python
df1 = pd.DataFrame({'이름':['영희','철수','민지','길동'],
                  '반':[1,2,3,4]})
df2 = pd.DataFrame({'이름':['광수','철수','민지','길동'],
                  '등급':[1,1,3,3],
                   '출생연도':['00','01','02','03']})
```

```python
pd.merge(df1, df2, on='이름',how='inner')
```

```python
#left
pd.merge(df1,df2, on='이름',how='left')
```

```python
#right
pd.merge(df1,df2, on='이름',how='right')
```

- inner만 NaN 값이 없고 나머지는 NaN 나옴!

```python
#outer 
pd.merge(df1,df2, on='이름',how='outer') #우리가 보고싶은 모든 이름이 다 나옴!
```

- inner를 제외한 나머지 병합에 대해서는 na 나올 수 있으니 꼭 유의하고 데이터프레임을 결합해야 한다!
