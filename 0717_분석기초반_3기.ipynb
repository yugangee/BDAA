## lambda 이용
- 반복 되는 데이터프레임을 간단하게 lambda 식을 이용해서 파생 변수를 만든다.

```python
exam.assign(new= exam['math']+exam['science']+exam['english'])
```

```python
exam.assign(new = lambda x: x['math']+x['science']+x['english'])
```

```python
exam.assign(new= exam['math']+exam['science']+exam['english'],
           mean = lambda x: x['new']/3) #람다를 이용해서 평균을 계산
```

## 집단별 요약

- 특정 컬럼을 기준으로 한 평균, 특정 컬럼을 기준으로한 전체 합 
- 내가 보고싶은 컬럼에 대해서 집단을 만들고 값을 계산

- groupby 

- agg - 기초통계값을 계산할 수 있는 함수

```python
exam
```

```python
exam.groupby('nclass').agg(mean_math = ('math', 'mean'))
```

```python
exam
```

- mean(평균)
- sum(합)
- max(최대)
- min(최소)
- std(표준편차)
- count(빈도, 개수)
- median(중위값)

```python
exam.groupby('nclass').agg(mean_math = ('math','mean'),
                          sum_math= ('math','std'),
                          median_math = ('math','median'),
                          n =('nclass', 'count'))
```

## 데이터 합치기

```python
test1 = pd.DataFrame({'id' :[1,2,3,4,5],
                     'midterm':[10,20,30,40,50]})
```

```python
test2 = pd.DataFrame({'id':[1,2,3,4,5],
                     'final':[60,80,40,20,10]})
```

```python
test1
```

```python
test2
```

```python
#데이터를 하나로 합치는 개념
pd.merge(test1, test2, how='left', on='id') 
```

```python
#데이터를 덩어리끼리 붙이는 경우
pd.concat([test1,test2])
```

## 이상치와 결측치

- 결측치 누락된 값, 비어있는 값, 이상치랑은 다른 내용

```python
df = pd.DataFrame({'a':['1','2', np.nan,'3','4'],
                  'b':['5','6','7',np.nan,'8']})
```

```python
df
```

```python
df.isna().sum()
```

```python
#결측치를 제거하는 경우
df.dropna(subset=['a'])
```

```python
#두 개의 변수에 대해서 na값을 제거

df.dropna(subset=['a','b'])
```

```python
df.dropna() #전체에 대한 결측치를 제거하는 방법
```
