```python
import pandas as pd
```

```python
import numpy as np
```

```python
mpg = pd.read_csv('mpg.csv')
```

```python
mpg
```

```python
mpg.assign(mpg = np.where(mpg['year']>=2005,'recent','old'))
```

```python
mpg.groupby('manufacturer').agg(mean_displ = ('displ', 'mean'))
```

```python
mpg.groupby('trans').agg(mean_displ = ('displ', 'mean'))
```

```python
mpg.groupby('manufacturer').mean()
```

```python
mpg.groupby(['manufacturer','model']).agg(mean_cty =('cty','mean'))
```

```python
mpg.query('manufacturer=="ford"').groupby(['drv']).agg(n =('drv','count'))
```
